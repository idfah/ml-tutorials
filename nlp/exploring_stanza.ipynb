```python
import time

import stanza
```

```python
with open('dissertation.txt', mode='r', encoding='utf8') as fh:
    text = fh.read()
```

```python
print(text[:1024])
```

```python
len(text)
```

```python
stanza.download('en')
```

```python
help(stanza.Pipeline)
```

```python
pipeline = stanza.Pipeline('en', processors='tokenize,ner')
```

```python
start_time = time.time()

document = pipeline(text)

duration = time.time() - start_time

print('took:', duration)
```
